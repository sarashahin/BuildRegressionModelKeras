```python
import pandas as pd
import numpy as np
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
#  Let's check how many data points we have
concrete_data.shape
```

```python
# missing values
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

```python
target.head()
```

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
n_cols = predictors_norm.shape[1] # number of predictors
```

```python
# Import Keras
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Build a Neural Network


# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# build the model
model = regression_model()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 1. Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the train_test_splithelper function from Scikit-learn.

X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
```

Train and Test the Network

```python
# build the model
model = regression_model()

# fit the model
model.fit(X_train, y_train, epochs=50, verbose=1)




```

```python
# evaluate the model
eval = model.evaluate(X_test, y_test, verbose=0)
eval
```

```python
# predict model

yprediction = model.predict(X_test)

```

```python
#  compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn.

from sklearn.metrics import mean_squared_error
```

```python
mean_squared = mean_squared_error(y_test, yprediction)
mean = np.mean(mean_squared)
std = np.std(mean_squared)

print("mean :" , mean, "std:" , std)
```

```python
# Repeat steps 1 - 3, 50 times, i.e., create a list of 50 mean squared errors.
tot_mean_squared_errors = 50
epochs= 50
mean_squared_errors = []
for i in range(0, tot_mean_squared_errors):
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=i)
    model.fit(X_train, y_train, epochs=epochs, verbose=0)
    eval1 = model.evaluate(X_test, y_test, verbose=0)
    print("eval1 "+str(i+1)+": "+str(eval1))
    yprediction = model.predict(X_test)
    mean_squared = mean_squared_error(y_test, yprediction)
    mean_squared_errors.append(mean_squared)
    
mean_squared_errors = np.array(mean_squared_errors)
mean = np.mean(mean_squared_errors)
std = np.std(mean_squared_errors)

print('\n')
print("Below is the mean and standard deviation of " +str(tot_mean_squared_errors) + " mean squared errors without normalized data. Total number of epochs for each training is: " +str(epochs) + "\n")
print("Mean: "+str(mean))
print("Standard Deviation: "+str(std))
    
    
```
